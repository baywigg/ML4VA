```python
import pandas as pd
import folium
```

```python
df = pd.read_csv("../data/clean/calls_general.csv")
```

```python
# Define the center of the map
map_center = [39.17131640406361, -78.16412213367745]  

# Create a folium map
m = folium.Map(location=map_center, zoom_start=12, min_zoom=12)

# Add markers to the map
for i in range(0, len(df), 100):
    row = df.iloc[i]
    folium.Marker([row["Latitude"], row["Longitude"]], popup=row["CallDescription"]).add_to(m)

# Display the map
m
```
